```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (np.sin(5*x))**2 / (2.5+(np.cos(2*x))) 
    
def generate(x, sigma=0.1):
    return f(x) + np.random.normal(0, sigma, len(x))
```

```python
np.random.seed(42)
n = 100
x = np.random.uniform(0, 1, n)
```

```python
plt.subplot(211)
plt.plot(x, f(x), '.')

plt.subplot(212)
plt.plot(x, generate(x), '.')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


np.random.seed(42)

X = x.reshape(-1, 1)
y_true = generate(x)
X_train, X_test, y_train, y_test = train_test_split(X, y_true, test_size=0.3)

x_plot = np.linspace(0, 1, 1000).reshape(-1, 1)


model_1 = make_pipeline(PolynomialFeatures(5), LinearRegression())
model_1.fit(X_train, y_train)
f_plot_1 = model_1.predict(x_plot)
mse_1 = mean_squared_error(y_test, model_1.predict(X_test))


model_2 = make_pipeline(PolynomialFeatures(20), LinearRegression())
model_2.fit(X_train, y_train)
f_plot_2 = model_2.predict(x_plot)
mse_2 = mean_squared_error(y_test, model_2.predict(X_test))


plt.plot(x_plot, f_plot_1, color='yellowgreen', linewidth=2,
         label="degree %d" % 5)
plt.plot(x_plot, f_plot_2, color='gold', linewidth=2,
         label="degree %d" % 20)

plt.plot(X_train, y_train, '.')
plt.legend(loc='lower left');
```

```python
print(f'deg = 5 => mse = {mse_1}, R2 = {model_1.score(X_train, y_train)}')

print(f'deg = 20 => mse = {mse_2}, R2 = {model_2.score(X_train, y_train)}')
```

```python
from tqdm import tqdm
n = 1000
preds = np.zeros((2, x.shape[0], n))
mse_history = np.zeros((2, n))
X = x[:, None]

for i in tqdm(range(n), total=n):
    y = generate(x)
    
    model_1.fit(X, y)
    model_2.fit(X, y)
    
    preds[0, :, i] = model_1.predict(X)
    preds[1, :, i] = model_2.predict(X)
    
    mse_history[0, i] = mean_squared_error(y, preds[0, :, i]) 
    mse_history[1, i] = mean_squared_error(y, preds[1, :, i]) 
    
```

```python
biass2 = ((preds - f(x)[None, :, None]).mean(axis=2) ** 2).mean(axis=1)
varience = ((preds - preds.mean(axis=2)[:, :, None]) ** 2).mean((1, 2))
eps = mse_history.mean(axis=1) - varience - biass2
for d, b, v, e in zip([5, 20], biass2, varience, eps):
    print(f'deg = {d}: \nbiass^2 = {b}, \nvarience = {v}, \nirreducible = {e}\n')
```
