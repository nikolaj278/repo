```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(5 * x) ** 2 / (2.5 + np.cos(2 * x))
    
def generate(x, sigma=0.1):
    return f(x) + np.random.normal(0, sigma, len(x))
```

```python
np.random.seed(42)
n = 100
x = np.linspace(0, 1, n)
```

```python

fig, axs = plt.subplots(2)

axs[0].plot(x, f(x), '-');
axs[0].plot(x, generate(x, sigma=0.05), '.');

axs[1].plot(x, f(x), '-');
axs[1].plot(x, generate(x, sigma=0.5), '.')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


np.random.seed(42)

X = x.reshape(-1, 1)
fig, axs = plt.subplots(2)

for i, sigma in enumerate([0.05, 0.5]):
    y_true = generate(x, sigma=sigma)
    X_train, X_test, y_train, y_test = train_test_split(X, y_true, test_size=0.3)
    
    x_plot = np.linspace(0, 1, 1000).reshape(-1, 1)
    
    model_1 = make_pipeline(PolynomialFeatures(2), LinearRegression())
    model_1.fit(X_train, y_train)
    f_plot_1 = model_1.predict(x_plot)
    mse_1 = mean_squared_error(y_test, model_1.predict(X_test))
    
    model_2 = make_pipeline(PolynomialFeatures(5), LinearRegression())
    model_2.fit(X_train, y_train)
    f_plot_2 = model_2.predict(x_plot)
    mse_2 = mean_squared_error(y_test, model_2.predict(X_test))
    
    
    model_3 = make_pipeline(PolynomialFeatures(20), LinearRegression())
    model_3.fit(X_train, y_train)
    f_plot_3 = model_3.predict(x_plot)
    mse_3 = mean_squared_error(y_test, model_3.predict(X_test))
    
    axs[i].set_title(f'sigma = {sigma}')
    axs[i].plot(x_plot, f_plot_1, color='pink', linewidth=2,
             label="degree %d" % 2)
    axs[i].plot(x_plot, f_plot_2, color='yellowgreen', linewidth=2,
             label="degree %d" % 5)
    axs[i].plot(x_plot, f_plot_3, color='gold', linewidth=2,
             label="degree %d" % 20)
    
    axs[i].plot(X_train, y_train, '.')
    axs[i].legend(loc='lower left');
```

```python
S = 2
D = 3
R = 300

mse_hist = np.zeros((S, D, R))
# models = [[] for _ in range(s)]

for i, sigma in enumerate([0.05, 0.5]):
    for j in range(R):
        y = generate(x)
        X_train, X_test, y_train, y_test = train_test_split(x.reshape(-1, 1), y, test_size=0.3)
        for k, degree in enumerate([2, 5, 20]):
            model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
            model.fit(X_train, y_train)
            # models[i].append(model)
            mse_hist[i, k, j] = mean_squared_error(y_test, model.predict(X_test))
            
    print(f'sigma = {sigma}, \nmeans are = {mse_hist[i].mean(axis=-1)}, \nstds are = {mse_hist[i].std(axis=-1)}\n')
```

```python
x_ = np.linspace(0, 1, n)
X = x_.reshape(-1, 1)
preds = np.zeros((S, D, R, n))

for i, sigma in enumerate([0.05, 0.5]):
    for j in range(R):
        y = generate(x_)
        for k, degree in enumerate([2, 5, 20]):
            model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
            model.fit(X, y)
            preds[i, k, j] = model.predict(X)
            mse_hist[i, k, j] = mean_squared_error(y, model.predict(X))

biass2 = ((f(x_)[None, None, None, :] - preds).mean(axis=3) ** 2).mean(axis=2)
varience = ((preds - preds.mean(axis=3)[:, :, :, None]) ** 2).mean((2, 3))
# eps = mse_hist.mean(axis=2) - varience - biass2
for i, sigma in enumerate([0.05, 0.5]):
    print(f'sigma = {sigma}')
    for d, b, v in zip([2, 5, 20], biass2[i], varience[i]):
        print(f'deg = {d}: \nbiass^2 = {b}, \nvarience = {v}')
    print('\n')
```
