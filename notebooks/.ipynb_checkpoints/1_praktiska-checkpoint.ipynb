```python
import sklearn
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
```

```python
def sample(n):
    np.random.seed(42)
    x0 = np.random.uniform(1, 10000, n)
    v0 = np.random.uniform(1, 100, n)
    t = np.random.uniform(1, 100, n)
    mass = np.random.uniform(1, 100, n)
    y = x0 + v0 * t - 9.81 * t**2 / 2
    return y, x0, v0, t, mass
```

```python
y, x0, v0, t, mass = sample(1000)
```

```python
X = np.vstack([x0, v0*t, t**2, mass]).T
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
lr = LinearRegression(n_jobs=-1)
lr.fit(X_train, y_train)
```

```python
preds = lr.predict(X_test)
```

```python
mse_test = mean_squared_error(y_test, preds)
mse_train = mean_squared_error(y_train, lr.predict(X_train))

print(f'mse test = {mse_test}, mse train = {mse_train}')
```

```python
import matplotlib.pyplot as plt

plt.plot(preds, y_test, 'o');
```

```python
lr.score(X_test, y_test) # R2 score
```

```python
import pandas as pd

print(*[f'{name} coef = {val}' for val, name in zip(lr.coef_, ['x0', 'v0*t', 't^2', 'mass'])], sep='\n')
```

The variable mass provides almost no information about the target variable because its coefficient is very small.

```python
g = -2 * lr.coef_[2]
print(f'g aproximation with the model is {g}')
```
